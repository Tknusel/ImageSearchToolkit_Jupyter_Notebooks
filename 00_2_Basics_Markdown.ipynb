# Markdown Basics

**Markdown** is a lightweight markup language that lets you format plain text using simple symbols. It was created by John Gruber in 2004 and is widely used in Jupyter notebooks, README files, documentation, and web content.

In Jupyter, cells can be set to **Markdown mode** (instead of Code mode) to render formatted text. Double-click any Markdown cell to see the raw syntax — press `Shift + Enter` to render it again.

---
## 1. Headings

Headings are created with `#` symbols. The number of `#` signs determines the level:

```
# Heading 1
## Heading 2
### Heading 3
#### Heading 4
```

Which renders as:

# Heading 1
## Heading 2
### Heading 3
#### Heading 4

---
## 2. Text Formatting

| Effect | Syntax | Result |
|---|---|---|
| Bold | `**text**` or `__text__` | **bold** |
| Italic | `*text*` or `_text_` | *italic* |
| Bold + Italic | `***text***` | ***bold italic*** |
| Strikethrough | `~~text~~` | ~~strikethrough~~ |

Example sentence:

```
This is **important**, this is *emphasis*, and this is ~~old~~.
```

Renders as:

This is **important**, this is *emphasis*, and this is ~~old~~.

## 3. Lists

### Unordered Lists
Use `-`, `*`, or `+` followed by a space:

```
- Apples
- Oranges
  - Navel
  - Blood orange
- Bananas
```

- Apples
- Oranges
  - Navel
  - Blood orange
- Bananas

### Ordered Lists
Use numbers followed by a period:

```
1. First step
2. Second step
3. Third step
```

1. First step
2. Second step
3. Third step

## 4. Links and Images

### Links
```
[link text](https://example.com)
```
[Visit the Jupyter website](https://jupyter.org)

### Images
Images use the same syntax as links, but with a `!` prefix:
```
![alt text](https://en.wikipedia.org/wiki/Indian_python#/media/File:Pratik_jain_dahod_python.JPG)
```

The **alt text** is shown if the image fails to load, and is important for accessibility.

## 5. Code

### Inline Code
Wrap code in single backticks: `` `print("Hello")` ``

Renders as: `print("Hello")`

### Code Blocks
Wrap in triple backticks. You can optionally specify a language for syntax highlighting:

````
```python
for i in range(5):
    print(f"Step {i}")
```
````

Renders as:

```python
for i in range(5):
    print(f"Step {i}")
```

## 6. Blockquotes

Use `>` at the start of a line:

```
> The purpose of computing is insight, not numbers.
> — Richard Hamming
```

> The purpose of computing is insight, not numbers.  
> — Richard Hamming

Blockquotes can be nested:

```
> Outer quote
> > Inner quote
```

> Outer quote
> > Inner quote

---
## 7. Tables

Tables use `|` as column separators and `-` for the header separator:

```
| Column A | Column B | Column C |
|----------|:--------:|---------:|
| Left     | Center   | Right    |
| aligned  | aligned  | aligned  |
```

| Column A | Column B | Column C |
|----------|:--------:|---------:|
| Left     | Center   | Right    |
| aligned  | aligned  | aligned  |

Alignment is controlled by the colon `:` position in the separator row:
- `---` → left (default)
- `:---:` → center
- `---:` → right

---
## 8. Horizontal Rules and Line Breaks

A **horizontal rule** is created with three or more dashes, asterisks, or underscores on their own line:

```
---
```

---

For a **line break** within a paragraph (without starting a new paragraph), end the line with **two spaces** before pressing Enter, or use `<br>`:

```
Line one  
Line two (same paragraph, new line)
```

Line one  
Line two (same paragraph, new line)

## 9. Inline HTML

Markdown supports raw HTML for things that Markdown can't do natively:

```html
<span style="color: steelblue; font-size: 1.2em;">Coloured text</span>

<details>
  <summary>Click to expand</summary>
  Hidden content goes here.
</details>
```

<span style="color: steelblue; font-size: 1.2em;">Coloured text</span>

<details>
  <summary>Click to expand</summary>
  Hidden content goes here.
</details>
