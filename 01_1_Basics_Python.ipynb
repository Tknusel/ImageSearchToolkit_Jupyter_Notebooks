## 0. Working in JupyterLab

A Jupyter notebook is a document made of **cells**. There are two main kinds:

- **Code cells** contain Python code. Run them to execute the code. The result appears directly below the cell.
- **Markdown cells** contain formatted text (like this one), used for notes, explanations, headings, links, and math.

You can change a cell's type via the dropdown in the toolbar, or with the keyboard shortcuts `Y` (code) and `M` (markdown) when the cell is selected but not being edited.

### Running cells

- `Shift` + `Enter` — run the current cell and move to the next one
- `Ctrl` + `Enter` (`Cmd` + `Enter` on macOS) — run the current cell and stay on it
- `Alt` + `Enter` — run and insert a new cell below

### Two modes

- **Edit mode** (green border in classic themes): you are typing inside a cell. Press `Esc` to leave.
- **Command mode** (blue border): the cell is selected but you are not editing it. Press `Enter` to go back to edit mode. Most single-letter shortcuts only work in command mode.

Useful command-mode shortcuts: `A` insert cell above, `B` insert below, `D D` delete, `Z` undo delete, `Shift` + `M` merge cells.

### The kernel

Every notebook is connected to a **kernel** — a running Python process that holds the state of your variables. If you restart the kernel, all variables are cleared. If something seems stuck or behaves strangely, `Kernel → Restart Kernel` is usually the fix. `Kernel → Restart and Run All` is useful to verify that your notebook actually works top-to-bottom.

Cells are not required to run in order, but your results will only be reproducible if they do. When in doubt: restart and run all.

## 3. Simple data types

While the technical reality is a bit more complicated, we can regard variables as *signs that hold values*. In Python, we assign a value to a sign with the equality operator: `a = 1`. Variables can hold many different kinds of values: integers (`a = 1`), floating point numbers (`a = 1.5`), strings (`a = 'foo'`), truth values (`True` or `False`), vectors, matrices, or complicated structures made up of many different kinds of values. Python also offers built-in support for more complex data types, for instance complex numbers (`a = 3+2j`) or very large/small numbers in scientific notation (`a = 2e-10`). Python is a [weakly-typed](https://en.wikipedia.org/wiki/Type_system) language — in practice, you do not have to tell the interpreter what kind of value you are assigning; it will infer it.

There are many ways to "look into" a variable. The easiest is to print its value with the `print` function: `print(a)`. In a notebook, if the *last line* of a cell is just an expression (e.g. `a` on its own), its value is also displayed automatically.

````python
a = 1 # Int
b = 1.0 # Float
c = "fghfghfghxfhgh" # String
# There are many more that we won't need
````

As variables hold values, we can use them algebraically, almost exactly like we would without a computer. For instance, we can calculate the sum of `a=4` and `b=2` by assigning it to `c`: `c = a + b`. Printing `c` will give us `6`. Many other algebraic operations are possible.

````python
a = 4
b = 2

print(a + b)
print(a - b)
print(a*b)  # Multiplication
print(a**b) # Power
print(a/b)  # Division always returns a floating point number
print(a%b)  # Modulo (remainder)
# There are many more that we won't need
````

## 4. Truth values

We can also obtain truth values from algebraic statements. To check for equality, we use the `==` operator:

````python
print(a == b)
print(a > b)
print(a < b)
print(a >= b)
print(a <= b)
````

## 5. Lists

One of the most convenient things in Python is its handling of lists, a compound data type. You can put almost anything between square brackets and pass it to a function to process. Items in a list usually have the same type, although they can also be of different types. Here are lists of integers and strings, respectively.

````python
int_list = [1, 2, 3, 4, 5]
string_list = ["foo", "bar", "baz"]
mixed_list = [1, 2, 5.9, "hello"]
print(int_list)
print(string_list)
print(mixed_list)
````

The most important feature of lists and similar compound data types — one that we will encounter over and over — is *slicing*. Slicing a list means returning just a part of it. This is done with the colon operator. For instance, to return just the first element of our integer list, we write `int_list[0]` (in programming, we almost always count from 0, not 1). The second element is `int_list[1]`, etc. We can also access parts of lists in a more dynamic fashion:

````python
print(int_list[0])  # The first element
print(int_list[:2]) # 1 2, the "first two" elements
print(int_list[2:]) # 3 4 5, everything but the "first two" elements
print(int_list[-2]) # 4, the "second to last" element
````

Lists are *mutable* — we can replace arbitrary elements by assigning a new value to them:

````python
int_list[0] = 100
print(int_list)
````

We can also concatenate lists:

````python
new_int_list = int_list[:2] + int_list[-2:]
print(new_int_list)
````

Note that for these examples we do not need to know how long the list is, or what is in it. If we *do* want to know how long a list is, we can use `len`: `len(new_int_list)` returns `4`. Slicing becomes more complicated but also more effective when we deal with multi-dimensional data types.

A word of caution: single items from lists are treated as simple types, not single-item lists. Concretely, `int_list[2] + int_list[-2]` does not return a new list but a value. To avoid this, we can either add extra brackets when dealing with single-item lists (`[int_list[2]] + [int_list[-2]]`) or use `append` to append items to a list:

````python
# Be careful:
print(int_list[2] + int_list[-2])

# Instead:
print([int_list[2]] + [int_list[-2]])

# Or:
int_list.append(6)
print(int_list)
````

## 6. Strings

It might not seem intuitive, but strings are really complicated to deal with for computers. The reasons are technical *and* historical, but it comes down to the fact that letters need to be represented as numbers. Check out the [Wikipedia entry for "Unicode"](https://en.wikipedia.org/wiki/Unicode) for a glimpse of the complexities of string representations. For us, one thing is important: strings behave very much like lists of letters. The good and bad of list indexing and slicing applies to strings as well.

````python
a = "I am a sentence."
print(a)
print(a[4])
print(a[0:4])
````

Python offers some very useful helper functions that you will use often when handling text:

````python
a = "    Apples, bananas, Cucumbers "

# Stripping spaces from a string
print(a.strip())

# Splitting a string at a specific separator
print(a.split(",")) # Returns a list, of course!

# Combining both with list comprehension
print([x.strip() for x in a.split(",")])

# Making a string out of a list
a_split = [x.strip() for x in a.split(",")]
print(" ".join(a_split)) # Use space as new separator

# Only lowercase
print(a.lower())

# Replace
print(a.replace("Apples", "Pears"))
````

## 7. Flow control: if…(then)…else

To do anything productive you will need to control the flow of your program — i.e. control what is executed when, not in terms of time but in terms of order. In most cases, flow control decisions are tied to the value of a variable. The simplest flow control measure is to have a piece of code execute conditionally on the value of a variable. In Python, we use `if … (then) … else`:

````python
a = 2
if (a == 0):
    print('A is zero')
elif (a == 1):
    print('A is one')
elif (a == 2):
    print('A is two')
else:
    print('A is neither zero nor one nor two')
````

Note the colon dividing the statement into "cause" and "effect". `if`-statements can be arbitrarily complex, for instance containing arguments chained through Boolean operators. In Python, Boolean operators are real words:

````python
a = 0
b = 1
if (a == 0 and b == 1):
    print('A is zero, b is one')
elif (a == 1 or b == 1):
    print('A or b or both are one')
````

All standard Boolean operators are available; many more can be obtained through libraries (and you can also build them by stacking standard operators).

## 8. Flow control: for

If you are used to another programming language, the `for` loop in Python takes some getting used to but is actually much more convenient than, for instance, its C or C++ cousin. In Python, a `for` loop always processes a list — technically it is a `for each` loop. The basic function of a `for` loop is to run some code a number of times until a condition is met, usually until a predefined number of times has been reached. To print "Hello" ten times, we write:

````python
for x in range(10):
    print("Hello")
````

Note that `range` here defines a generator that produces a sequence of numbers. We will not cover generators in detail; for now, it is sufficient to know that generators produce values one at a time.

````python
print(range(10))
print(list(range(10)))
````

If `range` is given one argument, it creates a sequence from 0 up to (but not including) the argument. With two arguments, it treats them as the lowest and highest values. A third argument defines the step size.

````python
for x in range(0, 20, 3):
    print(x)
````

Note the indentation: this is how Python knows we are defining a new code block that should run "inside" the `for` loop. The same indentation is used for `if`-statements.

As `for` loops always iterate over sequences, we can very easily iterate over the elements of an integer list:

````python
int_list = list(range(10))
for x in int_list:
    print(x)
````

To get both the index and the value of a list item, we can use the `enumerate` function:

````python
string_list = ["foo", "bar", "baz"]
for index, value in enumerate(string_list):
    print(index, value)
````

## 9. Flow control: while

An alternative to the `for` loop is the `while` loop. It continues running until a specified condition is met; unlike `for`, it does not operate on lists. The simplest `while` loop is an infinite loop (**do not run this** — or if you do, interrupt the kernel with the ■ stop button in the toolbar, or `Kernel → Interrupt Kernel`):

A more useful variation checks the value of a variable:

````python
a = 0
while a <= 10:
    print(a)
    a += 1
````

## 10. Functions

Functions in Python mirror mathematical functions. They are defined as an operation over several arguments. In Python, the `def` keyword is used, followed by the name, arguments in parentheses, a colon, and the indented function body. Functions can take anything between zero and arbitrarily many arguments. Arguments can be anything, even lists or other functions. Importantly, functions open up a new variable *scope*: variables defined within the function do not exist outside it, including the arguments. Results computed by a function are returned with the `return` statement.

````python
def add(a, b):
    result = a + b
    return result

c = add(1, 2)
print(c)
````

````python
# Variable scopes are important!
c = 0
def add(a, b):
    c = 100
    print('Inside function', c)
    return a + b

add(1, 2) # Note that this is just called and does not print the return value!
print('Outside function', c)
````

In Python, arguments can have default values. This is used extensively in machine learning code. Default values make arguments optional.

````python
def add(a, b=100):
    return a + b

print(add(1, 2))
print(add(1))
````

````python
# Arguments can be anything, even functions!

def add(a, b):
    return a + b

def subtract(a, b): # We can re-use argument names in a new scope
    return a - b

def arbitrary(f, a, b):
    return f(a, b)

print(arbitrary(add, 1, 2))
print(arbitrary(subtract, 1, 2))
print(arbitrary(print, 1, 2))
````

## 11. Dictionaries

Dictionaries extend lists in a useful way: they introduce key–value pairs which allow indexing an item by name, rather than by position. Dictionaries in Python use curly brackets, keys are typically strings, and values follow the key separated by a colon:

````python
empty_dict = {}
fruit = {'apples': 1, 'bananas': 2}
````

We can get an item from the dictionary like this:

````python
number_of_apples = fruit['apples']
print(number_of_apples)
````

To loop over a dictionary, Python 3 offers the `items()` method. For instance:

````python
for k, v in fruit.items():
    print(k, v)
````

will print the contents of the dictionary. Keys and values can be accessed separately. Values of keys can simply be overwritten:

````python
fruit['apples'] = 15
````

To check whether a single key is in the dictionary, use the `in` keyword:

````python
if 'apples' in fruit:
    print('We have', fruit['apples'], 'apples.')
````
